# <center>Clean And Analyze Employee Exit Surveys</center>

This project will be demonstrating data cleaning an analysis techniques through the analysis of employee exit surveys. The data used for analysis came from the [Department of Education, Training and Employment](https://en.wikipedia.org/wiki/Department_of_Education_and_Training_(Queensland) (DETE) and the Technical and Further Education (TAFE) institute in Queensland, Australia. Stakeholders can find the TAFE exit survey [here](https://data.gov.au/dataset/ds-qld-89970a3b-182b-41ea-aea2-6f9f17b5907e/details?q=exit%20survey) and the survey for the DETE [here](https://data.gov.au/dataset/ds-qld-89970a3b-182b-41ea-aea2-6f9f17b5907e/details?q=exit%20survey).

The following information will be presented to stakeholders:

* Are employees who only worked for the institutes for a short period of time resigning due to some kind dissatisfaction? What about employees who have been there longer?

* Are younger employees resigning due to some kind of dissatisfaction? What about older employees?

Below is a preview of columns being used from the `dete_survey.csv`:

* `ID`: An id used to identify the participant of the survey
* `SeperationType`: The reason why the person's employment ended
* `Cease Date`: The year or month the person's employment ended
* `DETE Start Date`: The year the person began employment with the DETE


Below is a preview of columns being used from the `tafe_survery.csv`:

* `Record ID`: An id used to identify the participant of the survey
* `Reason for ceasing employment`: The reason why the person's employment ended
* `LengthofServiceOverall..`: The length of the person's employment (in years)

## Verifying the Data

This step will be used to verify the validity of the data. It is necessary to make sure the data being used for analysis isn't corrupted in some way. 

The focus for the purposes of this analysis will be on verifying the years in the `cease_date` and `dete_start_date` columns correctly reflect what is expected for this dataset.

* Since the `cease_date` is the last year of the person's employment and the `dete_start_date` is the person's first year of employment it wouldn't make sense to have years after the current date.
* Given that most people in this field start working in their 20's, it's also unlikely that the `dete_start_date` was before the year 1940. 

If there are many responses with years higher than the current date or lower than 1940, analysis for the purpose this project would not be valid. If there are a small amount of values that are unrealistically high or low, they will be removed.

To analyze the date, the numbers will be converted into categories. This analysis is based on a [Businesswire](https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage), which makes the argument that understanding employee's needs according to career stage instead of age is more effective. 

The categories selected are listed are list below:
* New: less than 3 years at a company
* Experienced: 3-6 years at a company
* Established: 7-10 years at a company
* Veteran: 11 or more years at a company

```python
combined_updated['institute_service']
```

#### Extracting the years of service form each value in the `institute service` column:

```python
test = combined_updated['institute_service'].astype('str')
```

```python
test = test.str.replace('-','.')
```

```python
test = test.str.replace(' ', '.')
```

```python
test
```

```python
pattern = r"([0-9]+\.[0-9]?)"
```

```python
extracted = test.str.extract(pattern, expand = True)
```

```python
extracted = extracted.astype('float')
```

```python
extracted
```

```python
def mapper_func(val):
    if pd.isnull(val):
        return np.nan
    elif val < 3:
        return 'New: Less than 3 years at a company'
    elif val >= 3 and val <= 6:
        return 'Experienced: 3-6 years at a company'
    elif val >= 7 and val <= 10:
        return 'Established: 7-10 years at a company'
    else:
        return 'Veteran: 11 or more years at a company'
        
```

```python
combined_updated['service_cat'] = extracted.applymap(mapper_func)
```

```python
combined_updated['service_cat']
```
