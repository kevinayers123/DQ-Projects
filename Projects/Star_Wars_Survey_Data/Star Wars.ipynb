# <center>Analyzing Star Wars Survey Data</center>

While waiting for [Star Wars: The Force Awakens](https://en.wikipedia.org/wiki/Star_Wars:_The_Force_Awakens) to come out, the team at [FiveThirtyEight](http://fivethirtyeight.com) became interested in answering some questions about Star Wars fans. In particular, they wondered: <b>does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?</b>

The team needed to collect data addressing this question. To do this, they surveyed Star Wars fans using the online tool SurveyMonkey. They received 835 total responses, which can download from [their GitHub repository](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).

This project will be cleaning and exploring the Star Wars data set. The data has several columns, including:

* `RespondentID` - An anonymized ID for the respondent (person taking the survey)
* `Gender`- The respondent's gender
* `Age` - The respondent's age
* `Houshold Income` - The respondent's income
* `Education` - The respondent's education level
* `Location`(Census Region) - The respondent's location
* `Have you seen any of the 6 films in the Star Wars franchise?` - Has a `Yes` or `No` response
* `Do you consider yourself to be a fan of the Star Wars film franchise?` - Has a `Yes` or `No` response


## Reading the Data

```python
import pandas as pd
import numpy as np
from IPython.display import display

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

```python
star_wars.info()
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

## Cleaning and Mapping Yes/No Columns

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna = False)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna = False)
```

```python
yes_no = {"Yes": True, "NO": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

## Cleaning and Mapping Checkbox Columns

```python

import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

# Rankings

```python
for col in star_wars.columns[9:15]:
    star_wars[col] = star_wars[col].astype(float)
```

```python
star_wars[star_wars.columns[9:15]].info()
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

```python

star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

labels = star_wars[star_wars.columns[9:15]].mean().index.tolist()

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean(), align='center', alpha = 0.5)
plt.xticks(range(6), labels)
plt.ylabel('Average Movie Rank')
plt.show()
```

So far, the dataset has been cleaned, several columns have been renamed, and the average ranking of each move has been computed. As suspected, it looks like the "original" movies are rated much more highly than the new ones.

## View Counts

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
labels2 = star_wars[star_wars.columns[3:9]].sum().index.tolist()

plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum(), align='center', alpha = 0.5)
plt.xticks(range(6), labels2)
plt.ylabel('# Respondents Who Watched the Movie')
plt.show()
```

It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.

## Male/Female differences in favorite Star Wars movie and most seen movie

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
mens_label = males[males.columns[9:15]].mean().index.tolist()
womens_label = females[females.columns[9:15]].mean().index.tolist()
```

```python
plt.bar(range(6), males[males.columns[9:15]].mean(), align='center', alpha = 0.5)
plt.xticks(range(6), mens_label) 
plt.ylabel('Average Movie Rank Men')
plt.show()
```

```python
plt.bar(range(6), females[females.columns[9:15]].mean(), align='center', alpha = 0.5)
plt.xticks(range(6), womens_label) 
plt.ylabel('Average Movie Rank Women')
plt.show()
```

```python
mens_label2 = males[males.columns[3:9]].sum().index.tolist()
womens_label2 = females[females.columns[3:9]].sum().index.tolist()
```

```python
plt.bar(range(6), males[males.columns[9:15]].sum(), align='center', alpha = 0.5)
plt.xticks(range(6), mens_label2) 
plt.ylabel('# Male Respondents Who Watched the Movie')
plt.show()
```

```python
plt.bar(range(6), females[females.columns[9:15]].sum(), align='center', alpha = 0.5)
plt.xticks(range(6), womens_label2) 
plt.ylabel('# female Respondents Who Watched the Movie')
plt.show()
```

Interestingly, more males watched episodes 1-3, but males liked them far less than females did.
